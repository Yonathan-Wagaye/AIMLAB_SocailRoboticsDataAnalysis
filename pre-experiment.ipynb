```python
import pandas as pd
```

```python
file_path = './PreExperimentData.csv'
df = pd.read_csv(file_path)
```

```python
df = df.iloc[4:]
df = df.iloc[:, 2:]
```

```python
new_columns = { df.columns[i]: f'Q{i}' for i in range(18)}
```

```python
df = df.rename(columns = new_columns)
```

```python
df.head()
```

```python
response_mapping_one = {
    'Never': 0,
    'Rarely': 0,
    'Sometimes': 1,
    'Often': 1,
    'Very Often': 1
}

response_mapping_two = {
    'Never': 0,
    'Rarely': 0,
    'Sometimes': 0,
    'Often': 1,
    'Very Often': 1
}


def calculate_adhd_score(row):
    score = 0
    for q in ['Q0', 'Q1', 'Q2']:
        score += response_mapping_one[row[q]]
    for q in ['Q3', 'Q4', 'Q5']:
         score += response_mapping_two[row[q]]     
    return score
```

```python
df['Part_A_Score'] = df.apply(calculate_adhd_score, axis=1)
df['ADHD_Indication'] = df['Part_A_Score'] >= 4
```

```python
def calculate_percentage(df):       
    adhd_percentage = (df['ADHD_Indication'].sum() / len(df)) * 100
    return round(adhd_percentage,2)
```

```python
print(f"Possible ADHD: {calculate_percentage(df)}%")
```
